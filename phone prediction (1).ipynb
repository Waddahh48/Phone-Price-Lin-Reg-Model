```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
phone= pd.read_csv('phone.csv')
phone
```

```python
phone.isnull().sum()
```

```python
phone['Price(£)']=phone['Price']/93.55
phone.rename(columns = {'Price':'Price(INR)'}, inplace = True)
phone
```

```python
map={'Yes':1,'No':0}
list=['Touchscreen','Wi-Fi','Bluetooth','GPS','3G','4G/ LTE']
for i in list:
    phone[i]=phone[i].map(map)
phone
```

```python
phone['Operating system'].nunique()
```

```python
p = pd.get_dummies(phone,columns=['Operating system','Brand','Model'])
p
```

```python
p1=p.drop(['Name','Unnamed: 0','Price(INR)'],axis=1)
```

```python
x = p1.drop('Price(£)',axis=1)
y = p1['Price(£)']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=0)
reg = LinearRegression()
```

```python
reg.fit(X_train,y_train)
reg.score(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge(1)
ridge.fit(X_train,y_train)
ridge.score(X_train,y_train)
```

```python
ridge.score(X_test,y_test)
```
